# 一个代码列表长度500，要求以80个为一个批次，拆分成多个数组打印输出

```python
import csv
def read_csv_file(file_path):
    data = [] 
    with open(file_path, 'r') as f:  # 打开指定路径的CSV文件，以只读模式
        reader = csv.reader(
            f)
        for row in reader:  # 遍历
            data.append(row)  
    return data  # 返回存储了CSV文件数据的列表
file_path = 'C:/Users/Cyan_wind/Downloads/fyx_chinamoney.csv' 
data = read_csv_file(file_path)  
data
```

```python
code_list = data 
batch_size = 80 

split_list = []
for i in range(0, len(code_list), batch_size):
    split_list.append(code_list[i:i+batch_size])

for i, batch in enumerate(split_list):
    print(f"Batch {i+1}: {batch}")
```
